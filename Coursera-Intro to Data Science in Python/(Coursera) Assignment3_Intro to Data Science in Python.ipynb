### Question 7 (6.6%)

### Question 9 (6.6%)
Create a column that estimates the number of citable documents per person. 
What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the `.corr()` method, (Pearson's correlation).

*This function should return a single number.*

````python
def answer_nine():
    Top15 = answer_one()
    
    Top15['Population Estimate'] =\
    Top15.copy()['Energy Supply'] / Top15.copy()['Energy Supply per Capita']
    
    Top15['Citable documents per Capita']=\
    Top15.copy()['Citable documents'] / Top15.copy()['Population Estimate']
    
    return Top15.copy().loc[:,['Citable documents per Capita', 'Energy Supply per Capita']].corr().values[0,1]
````

### Question 10 (6.6%)
Create a new column with a 1 if the country's % Renewable value is at or above the median for all countries in the top 15, and a 0 if the country's % Renewable value is below the median.

*This function should return a series named `HighRenew` whose index is the country name sorted in ascending order of rank.*

````python
def renew_value_0_or_1(x):
    if x >= np.median(x):
        return 1
    else:
        return 0

def answer_ten():
    Top15 = answer_one()
    Top15['HighRenew'] =\
    Top15.copy()['% Renewable'].apply(renew_value_0_or_1)
    return Top15['HighRenew']
````

### Question 12 (6.6%)
Cut % Renewable into 5 bins. Group Top15 by the Continent, as well as these new % Renewable bins. How many countries are in each of these groups?

*This function should return a __Series__ with a MultiIndex of `Continent`, then the bins for `% Renewable`. Do not include groups with no countries.*

````python
def answer_twelve():
    Top15 = answer_one()
    
    Top15['Renew_bins']=\
    pd.cut(Top15.copy()['% Renewable'], 5)
    
    return Top15.copy().reset_index().groupby(['Continent', 'Renew_bins'])['Country'].count()
````
